```
# HOUSING PRICING PREDICTION USING CALIFORNIA HOUSING DATASET (Regression)

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
housing = fetch_california_housing(as_frame=True)
x = housing.data
y = housing.target

# Split data
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

# Training and prediction by Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
y_pre_lr = lin_reg.predict(x_test)

# Training and prediction by Decision Tree
tree_reg = DecisionTreeRegressor(max_depth=5, random_state=42)
tree_reg.fit(x_train, y_train)
y_pre_tree = tree_reg.predict(x_test)

# Evaluation on the test set
print("Evaluation on the test set BY Linear Regression")
mse_lr = mean_squared_error(y_test, y_pre_lr)
r2_lr = r2_score(y_test, y_pre_lr)
print("MSE: ", mse_lr)
print("R² Score: ", r2_lr)

print("\nEvaluation on the test set BY Decision Tree Regressor")
mse_tree = mean_squared_error(y_test, y_pre_tree)
r2_tree = r2_score(y_test, y_pre_tree)
print("MSE: ", mse_tree)
print("R² Score: ", r2_tree)

# Visualization (Predicted vs Actual)
plt.figure(figsize=(12, 5))

# Linear Regression
plt.subplot(1, 2, 1)
plt.scatter(y_test, y_pre_lr, alpha=0.5, color="blue", label="Predictions")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()],
         'r--', label="Perfect Prediction")
plt.xlabel("Actual House Prices ($100,000s)")
plt.ylabel("Predicted House Prices ($100,000s)")
plt.title("Linear Regression Predictions")
plt.legend()

# Decision Tree Regressor
plt.subplot(1, 2, 2)
plt.scatter(y_test, y_pre_tree, alpha=0.5, color="green", label="Predictions")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()],
         'r--', label="Perfect Prediction")
plt.xlabel("Actual House Prices ($100,000s)")
plt.ylabel("Predicted House Prices ($100,000s)")
plt.title("Decision Tree Predictions")
plt.legend()

# Show graphs
plt.tight_layout()
plt.show()

# Comparison of both models
results = pd.DataFrame({
    "Model": ["Linear Regression", "Decision Tree"],
    "MSE": [mse_lr, mse_tree],
    "R² Score": [r2_lr, r2_tree]
})

print("\nComparison Table:")
print(results)

# Observations
print("\nObservations:")
if r2_tree > r2_lr:
    print("Decision Tree performed better in terms of R² score (explains more variance).")
else:
    print("Linear Regression performed better in terms of R² score (explains more variance).")

if mse_tree < mse_lr:
    print("Decision Tree has lower prediction error (MSE), so Decision Tree Regressor is better.")
else:
    print("Linear Regression has lower prediction error (MSE), so Linear Regression is better.")
```
